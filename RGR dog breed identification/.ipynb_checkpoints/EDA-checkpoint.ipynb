```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from pathlib import Path
from PIL import Image
```

Prepare dataset

```python
images_folder = Path("dog-breed-identification")
train_images_folder = images_folder / "train"
train_resized_folder = images_folder / "train_resized"

test_images_folder = images_folder / "test"

labels_path = Path("dog-breed-identification/labels.csv")
```

```python
from skimage.io import imread
from skimage.transform import resize
from progress_bar import printProgressBar
from itertools import count


def get_images_from_folder(folder, printSuffix=''):
    
    total_files = len(list(folder.rglob("*.jpg")))

    for file_index, image_path in zip(count(1), folder.rglob("*.jpg")):
        yield (image_path.stem, imread(image_path))
        printProgressBar(file_index, total_files, prefix = "Reading images", suffix=printSuffix, printEnd='')

def get_resized_from_folder(folder, width=100, height=None):

    height = width if height is None else height
    new_size = (width, height)

    for file_name, image in get_images_from_folder(folder, f"resizing to {size}"):
        yield (file_name, resize(image, new_size))
        

def read_image_for_id(id, folder=train_images_folder):
    image_path = list(folder.glob(f"{id}.jpg"))[0]
    return Image.open(image_path)
```

```python
labels = data_labels['breed'].cat.categories[:10]
fig, axes = plt.subplots(1, 10)
fig.set_size_inches(20, 5)


for ax, label in zip(axes, labels):
    sample = data_labels[data_labels['breed'] == label].sample(1)
    
    ax.imshow(read_image_for_id(sample.iloc[0]['id']).resize((200, 200)))
    ax.axis('off')
    ax.set_title(label)
```
